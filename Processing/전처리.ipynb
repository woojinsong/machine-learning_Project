## 가족수 이상치 제거 (전처리 시 0도 제거 되는 듯...)

```python
plt.figure(figsize=(12,8))
sns.boxplot(data['familysize'], color='red')
plt.show()
```

```python
data['familysize'].unique()
```

```python
def outliers_iqr(data):
    q1, q3 = np.percentile(data, [25,75])
    iqr = q3 - q1
    lower_bound = q1 - (iqr * 1.5)
    upper_bound = q3 + (iqr * 1.5)
    
    return np.where((data>upper_bound)|(data<lower_bound))
```

```python
family_outlier_index = outliers_iqr(data['familysize'])[0]
```

```python
data.loc[family_outlier_index, 'familysize']
```

```python
family_not_outlier_index=[]

for i in data.index:
    if i not in family_outlier_index : 
        family_not_outlier_index.append(i)
```

```python
data_clean = data.loc[family_not_outlier_index]
data_clean = data_clean.reset_index(drop=True)
data_clean
```

```python
# data_clean.to_excel('전처리_마키아벨리니즘_가족이상치제거1.xlsx')
data_clean.to_excel('전처리_가족이상치제거1.xlsx')
```

```python
data_clean['familysize'].unique()
```

## 가족수 이상치 제거 (0 포함)

```python
def outliers_iqr(data):
    q1, q3 = np.percentile(data, [25,75])
    iqr = q3 - q1
    lower_bound = q1 - (iqr * 1.5)
    upper_bound = q3 + (iqr * 1.5)
    
    return np.where(data>upper_bound)

family_outlier_index = outliers_iqr(data['familysize'])[0]
```

```python
data.loc[family_outlier_index, 'familysize']
```

```python
family_not_outlier_index=[]

for i in data.index:
    if i not in family_outlier_index : 
        family_not_outlier_index.append(i)
```

```python
data_clean = data.loc[family_not_outlier_index]
data_clean = data_clean.reset_index(drop=True)
data_clean
```

```python
# data_clean.to_excel('전처리_마키아벨리니즘_가족이상치제거2.xlsx')
data_clean.to_excel('전처리_가족이상치제거2.xlsx')
```
